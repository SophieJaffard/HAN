```python
from hansolo import *
M_obj=500
M=2502
n_epoch = int(M/9)

n2=10

Nb=100

level=0.1
learning="blocs"

plt.figure(figsize=(4,3))

legend=['HAN with EWA', 'HAN with PWA', 'HAN Solo with EWA', 'HAN Solo with PWA', 'Component-Cue','Perceptron']
color=['green','magenta','blue','orange', 'brown','red']

#answers=np.load(f'saves/PWA_{learning}_HAN_test.npy')

answers=np.zeros((6,Nb,n2, M_obj))

answers[2,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/saves/EWA_{learning}_test.npy')[:,:n2,:]
answers[3,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/saves/PWA_{learning}_test.npy')[:,:n2,:]
answers[0,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/saves/EWA_{learning}_HAN_test.npy')[:,:n2,:]
answers[1,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/saves/PWA_{learning}_HAN_test.npy')[:,:n2,:]
answers[4,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/saves/{learning}_CC_test.npy')[:,:n2,:]
answers[5,:,:,:]=np.load(f'/Users/sophiejaffard/Desktop/HAN2/{learning}_percep_test.npy')[:,:n2,:]


for i in range(6):
    Y=np.zeros((Nb,n2)) 
    for nb in range(Nb):
        for m in range(n2):
            Y[nb,m]=np.sum(answers[i,nb,m,:])/M_obj
    Y_med=np.zeros(n2)
    Y_90=np.zeros(n2)
    Y_10=np.zeros(n2)

    for m in range(n2):
        X=Y[:,m]
        X=np.sort(X)
        Y_med[m]=X[int(Nb/2)]
        Y_90[m]=X[int(Nb*(1-level))]
        Y_10[m]=X[int(Nb*level)]
   

    X2=np.arange(n2)
    if i==4 or i==3:
        plt.plot(X2,Y_med,label=legend[i], color=color[i])
    else:
        plt.plot(X2,Y_med,label=legend[i], color=color[i])
    plt.fill_between(X2, Y_10, Y_90, color=color[i], alpha=.2)

#plt.yticks([0.6,0.8,1])
#plt.xticks([0,1500,2500])
plt.ylabel('Proportion of correct classifications')
plt.xlabel('Epoch')

plt.legend()
#plt.savefig('curves_percep.pdf')
plt.savefig("curves_percep.pdf", bbox_inches="tight")

plt.show()

```
